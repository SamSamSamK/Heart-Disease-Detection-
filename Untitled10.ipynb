```python
##Resting Blood-Pressure Distribution Overview
```

```python
df.columns
```

```python
df['trestbps'].hist()
```

```python
desc = df['trestbps'].describe()
print("Summary Statistics:\n", desc)

# Count how many people fall into clinically relevant categories
bp_categories = pd.cut(df['trestbps'],
                       bins=[0, 120, 130, 140, 180, float('inf')],
                       labels=['Normal (<120)', 'Elevated (120-129)', 'Hypertension Stage 1 (130-139)',
                               'Hypertension Stage 2 (140-179)', 'Hypertensive Crisis (180+)'])

category_counts = bp_categories.value_counts().sort_index()
print("\nResting Blood Pressure Categories:\n", category_counts)

# Print logical interpretation
print("\nInterpretation:")
print(f"- Mean resting blood pressure: {desc['mean']:.2f} mm Hg")
print(f"- Median resting blood pressure: {desc['50%']:.2f} mm Hg")
print(f"- Minimum: {desc['min']} mm Hg, Maximum: {desc['max']} mm Hg")
print(f"- Most people have blood pressure between 120–140 mm Hg.")

# Add category percentage interpretation
total = len(df)
for cat, count in category_counts.items():
    percent = (count / total) * 100
    print(f"- {cat}: {count} patients ({percent:.1f}%)")
```
